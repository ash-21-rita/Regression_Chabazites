```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import numpy as np


ac_cooling_kcha = pd.read_csv("C:\\Users\\Ashrita\\Downloads\\accooling.csv")  
perm_cooling_kcha = pd.read_csv("C:\\Users\\Ashrita\\Downloads\\permcooling.csv")  
ac_heating_kcha = pd.read_csv("C:\\Users\\Ashrita\\Downloads\\ACheating.csv")  
perm_heating_kcha = pd.read_csv("C:\\Users\\Ashrita\\Downloads\\permheating.csv")  

print("KCha AC Conductivity Cooling Data:\n", ac_cooling_kcha.head())
print("\nKCha AC Conductivity Heating Data:\n", ac_heating_kcha.head())
print("\nKCha Permittivity Cooling Data:\n", perm_cooling_kcha.head())
print("\nKCha Permittivity Heating Data:\n", perm_heating_kcha.head())

def plot_trends(data, y_columns, title, ylabel):
    plt.figure(figsize=(10, 6))
    for col in y_columns:
        plt.plot(data['Frequency / Hz'], data[col], label=col)
    plt.xscale('log')
    plt.xlabel('Frequency (Hz)')
    plt.ylabel(ylabel)
    plt.title(title)
    plt.legend()
    plt.grid(True, which='both', linestyle='--', linewidth=0.5)
    plt.show()

conductivity_cols_cooling = [col for col in ac_cooling_kcha.columns if 'Cond.' in col]
conductivity_cols_heating = [col for col in ac_heating_kcha.columns if 'Cond.' in col]

plot_trends(ac_cooling_kcha, conductivity_cols_cooling, 'KCha AC Conductivity Cooling Trends', 'Conductivity (S/m)')
plot_trends(ac_heating_kcha, conductivity_cols_heating, 'KCha AC Conductivity Heating Trends', 'Conductivity (S/m)')

permittivity_cols_cooling = [col for col in perm_cooling_kcha.columns if 'Perm' in col]
permittivity_cols_heating = [col for col in perm_heating_kcha.columns if 'Perm' in col]

plot_trends(perm_cooling_kcha, permittivity_cols_cooling, 'KCha Permittivity Cooling Trends', 'Permittivity')
plot_trends(perm_heating_kcha, permittivity_cols_heating, 'KCha Permittivity Heating Trends', 'Permittivity')

# Linear Regression Analysis
def regression_analysis(data, x_column, y_columns, ylabel):
    print(f"\nRegression Analysis for {ylabel}:\n")
    for col in y_columns:
        X = np.log10(data[x_column]).values.reshape(-1, 1)  # Log-scale frequency
        y = data[col].values
        model = LinearRegression()
        model.fit(X, y)
        y_pred = model.predict(X)
        
        # Calculate R^2 score
        r2 = r2_score(y, y_pred)
        
        print(f"{col}:")
        print(f"  Coefficients: {model.coef_}")
        print(f"  Intercept: {model.intercept_}")
        print(f"  R^2 Score: {r2}\n")
        
        plt.figure(figsize=(8, 5))
        plt.scatter(X, y, label='Observed', color='blue', alpha=0.6)
        plt.plot(X, y_pred, label='Fitted', color='red')
        plt.xlabel('Log10(Frequency)')
        plt.ylabel(ylabel)
        plt.title(f'Regression Analysis for {col}')
        plt.legend()
        plt.grid(True)
        plt.show()

regression_analysis(ac_cooling_kcha, 'Frequency / Hz', conductivity_cols_cooling, 'Conductivity (S/m)')

regression_analysis(ac_heating_kcha, 'Frequency / Hz', conductivity_cols_heating, 'Conductivity (S/m)')

regression_analysis(perm_cooling_kcha, 'Frequency / Hz', permittivity_cols_cooling, 'Permittivity')

regression_analysis(perm_heating_kcha, 'Frequency / Hz', permittivity_cols_heating, 'Permittivity')
```
