```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import numpy as np

# Load Cooling and Heating Data for KCha
ac_cooling_kcha = pd.read_csv("C:\\Users\\Ashrita\\Downloads\\accooling.csv")  # Replace with your KCha cooling conductivity file path
perm_cooling_kcha = pd.read_csv("C:\\Users\\Ashrita\\Downloads\\permcooling.csv")  # Replace with your KCha cooling permittivity file path
ac_heating_kcha = pd.read_csv("C:\\Users\\Ashrita\\Downloads\\ACheating.csv")  # Replace with your KCha heating conductivity file path
perm_heating_kcha = pd.read_csv("C:\\Users\\Ashrita\\Downloads\\permheating.csv")  # Replace with your KCha heating permittivity file path

# Display the data for verification
print("KCha AC Conductivity Cooling Data:\n", ac_cooling_kcha.head())
print("\nKCha AC Conductivity Heating Data:\n", ac_heating_kcha.head())
print("\nKCha Permittivity Cooling Data:\n", perm_cooling_kcha.head())
print("\nKCha Permittivity Heating Data:\n", perm_heating_kcha.head())

# Define a function to plot trends
def plot_trends(data, y_columns, title, ylabel):
    plt.figure(figsize=(10, 6))
    for col in y_columns:
        plt.plot(data['Frequency / Hz'], data[col], label=col)
    plt.xscale('log')
    plt.xlabel('Frequency (Hz)')
    plt.ylabel(ylabel)
    plt.title(title)
    plt.legend()
    plt.grid(True, which='both', linestyle='--', linewidth=0.5)
    plt.show()

# Plot Conductivity Trends for Cooling and Heating
conductivity_cols_cooling = [col for col in ac_cooling_kcha.columns if 'Cond.' in col]
conductivity_cols_heating = [col for col in ac_heating_kcha.columns if 'Cond.' in col]

plot_trends(ac_cooling_kcha, conductivity_cols_cooling, 'KCha AC Conductivity Cooling Trends', 'Conductivity (S/m)')
plot_trends(ac_heating_kcha, conductivity_cols_heating, 'KCha AC Conductivity Heating Trends', 'Conductivity (S/m)')

# Plot Permittivity Trends for Cooling and Heating
permittivity_cols_cooling = [col for col in perm_cooling_kcha.columns if 'Perm' in col]
permittivity_cols_heating = [col for col in perm_heating_kcha.columns if 'Perm' in col]

plot_trends(perm_cooling_kcha, permittivity_cols_cooling, 'KCha Permittivity Cooling Trends', 'Permittivity')
plot_trends(perm_heating_kcha, permittivity_cols_heating, 'KCha Permittivity Heating Trends', 'Permittivity')

# Linear Regression Analysis
def regression_analysis(data, x_column, y_columns, ylabel):
    print(f"\nRegression Analysis for {ylabel}:\n")
    for col in y_columns:
        # Prepare data
        X = np.log10(data[x_column]).values.reshape(-1, 1)  # Log-scale frequency
        y = data[col].values
        
        # Perform linear regression
        model = LinearRegression()
        model.fit(X, y)
        y_pred = model.predict(X)
        
        # Calculate R^2 score
        r2 = r2_score(y, y_pred)
        
        # Print results
        print(f"{col}:")
        print(f"  Coefficients: {model.coef_}")
        print(f"  Intercept: {model.intercept_}")
        print(f"  R^2 Score: {r2}\n")
        
        # Plot regression results
        plt.figure(figsize=(8, 5))
        plt.scatter(X, y, label='Observed', color='blue', alpha=0.6)
        plt.plot(X, y_pred, label='Fitted', color='red')
        plt.xlabel('Log10(Frequency)')
        plt.ylabel(ylabel)
        plt.title(f'Regression Analysis for {col}')
        plt.legend()
        plt.grid(True)
        plt.show()

# Regression for Conductivity Cooling (KCha)
regression_analysis(ac_cooling_kcha, 'Frequency / Hz', conductivity_cols_cooling, 'Conductivity (S/m)')

# Regression for Conductivity Heating (KCha)
regression_analysis(ac_heating_kcha, 'Frequency / Hz', conductivity_cols_heating, 'Conductivity (S/m)')

# Regression for Permittivity Cooling (KCha)
regression_analysis(perm_cooling_kcha, 'Frequency / Hz', permittivity_cols_cooling, 'Permittivity')

# Regression for Permittivity Heating (KCha)
regression_analysis(perm_heating_kcha, 'Frequency / Hz', permittivity_cols_heating, 'Permittivity')
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load Cooling and Heating Data for KCha
ac_cooling_kcha = pd.read_csv("C:\\Users\\Ashrita\\Downloads\\accooling.csv")  # Replace with your KCha cooling conductivity file path
perm_cooling_kcha = pd.read_csv("C:\\Users\\Ashrita\\Downloads\\permcooling.csv")  # Replace with your KCha cooling permittivity file path
ac_heating_kcha = pd.read_csv("C:\\Users\\Ashrita\\Downloads\\ACheating.csv")  # Replace with your KCha heating conductivity file path
perm_heating_kcha = pd.read_csv("C:\\Users\\Ashrita\\Downloads\\permheating.csv")  # Replace with your KCha heating permittivity file path
# Combine Cooling and Heating Data for Correlation Analysis
cooling_data = ac_cooling_kcha.merge(perm_cooling_kcha, on="Frequency / Hz", suffixes=('_Cond', '_Perm'))
heating_data = ac_heating_kcha.merge(perm_heating_kcha, on="Frequency / Hz", suffixes=('_Cond', '_Perm'))

# Function to Plot Correlation Heatmap
def plot_correlation_heatmap(data, title):
    correlation_matrix = data.corr()  # Calculate correlations
    plt.figure(figsize=(10, 8))
    sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", square=True)
    plt.title(title)
    plt.show()

# Plot Heatmaps for Cooling and Heating Data
plot_correlation_heatmap(cooling_data.drop(columns=['Frequency / Hz']), 'Correlation Heatmap: Cooling Data (KCha)')
plot_correlation_heatmap(heating_data.drop(columns=['Frequency / Hz']), 'Correlation Heatmap: Heating Data (KCha)')
```
